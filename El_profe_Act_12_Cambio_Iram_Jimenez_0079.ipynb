## Las variables
1. CP --> Costo del producto
2. CD --> Cantidad de dinero
3. Cambio

## Formulas
$Cambio = CD - CP$
## Diagrama de flujo Lucidchart
<img src="https://raw.githubusercontent.com/IramJimenez128/Calcular-cambio-if-else/refs/heads/main/Diagrama_flujo.png" width="500" height="600" alt="Descripción de la imagen">

```python
# Para cuando es verdadero (si)
# Calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# Si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Iram Jimenez NL = 28")
```

```python
# Para cuando es falso (no)
# Calculando el cambio usando if else
# Datos de entrada CD, CP
# Mostrar cambio si es falso
# Si es verdadero mostrar no alcanza
CP = float(input("Ingrese el costo del producto: "))
CD = float(input("Ingrese la cantidad de dinero: "))
if CD < CP:
    print("No alcanza")
else:
    Cambio = CD - CP
    print("El cambio es: ", Cambio)
print("Fin del programa")
print("Programa realizado por Iram Jimenez NL = 28")
```
